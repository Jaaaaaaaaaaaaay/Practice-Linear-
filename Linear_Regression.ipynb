Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

1. A Reals estate wishes to examine the relationship between the selling price of a home and its size

```python
#reading the house simple data set as a data frame
house_df = pd.read_csv('housesimple.csv')

#independent(x) and dependent variables(y)
X = np.array(house_df[['sqft']])
y = np.array(house_df[['price']])

#scatter plot for the relation of size and price
plt.scatter(X, y, color='green')
plt.ylabel('Price in $1000s')
plt.xlabel('Size in square feet')
plt.title('Size and Price Relation')
plt.show()
```

**Building the linear regression model**

```python
#splitting data into test and train
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

#linear regression as a model
model = LinearRegression()

#training the model
model.fit(X_train, y_train)

#testing the model
y_pred=model.predict(X_test)
```

**Comparison of Values between the predicted values and the actual values**

```python
comparison_df = pd.DataFrame({'Actual Values': y_test.flatten(), 'Predicted Values' : y_pred.flatten()})
print(comparison_df)

comparison_df.plot(kind='bar', figsize=(5,5))
plt.grid(linestyle='-', linewidth='0.5')
plt.grid(linestyle=':', linewidth='0.5')
plt.show()
```

**Scatter plot of the actual and predicted values of the linear regression model**

```python
plt.scatter(X_test, y_test, color='green')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.ylabel('Price')
plt.xlabel('Size')
plt.title('Prediction values and Actual Values')
plt.show()
```

2. Explain and apply Linear regression in the Players.csv

A quick introduction on how many players are playing on each position available before proceeding to the linear regression.

```python
players_df = pd.read_csv('Players.csv')
players_df.head()

playerfreq = players_df['position'].value_counts()
colors = []

playerfreq.plot(kind = 'bar')
plt.ylabel("Positions Played")
```

As plotted below, all of the details are presented as "The longer you play, equals to many of your contribution to the game". this is presented below and represented by each of the following dots that is scattered directly to show how many minutes and how many scores each player made. However all of this does not apply to everyone, as there are a lot of player that had more minutes of playtime but almost having no contribution, this excludes the goalkeeper as they are simply not choosing to to shoot and would rather play to save, and likewise as a lot will not be saving the ball as they are more likely to attack, defend and this attempt would be more likely done by a goalkeeper.

in summary, all of these data are useful for future references and how to monitor the playstyle of a player, either regular or professional. this can be used for products or for improvements or even for monitoring in the future.

```python
players_df = pd.read_csv('Players.csv')

minutes = np.array(players_df[['minutes']])
passes = np.array(players_df[['passes']])
shots = np.array(players_df[['shots']])
tackles = np.array(players_df[['tackles']])
saves = np.array(players_df[['saves']])

plt.figure(figsize=(10,6))

#passes plot
plt.subplot(2,2,1)
plt.scatter(minutes, passes, color='blue')
plt.ylabel('Passes')
plt.xlabel('Minutes')

#shots plot
plt.subplot(2,2,2)
plt.scatter(minutes, shots, color='green')
plt.ylabel('Shots')
plt.xlabel('Minutes')

#tackles plot
plt.subplot(2,2,3)
plt.scatter(minutes, tackles, color='purple')
plt.ylabel('Tackles')
plt.xlabel('Minutes')

#saves plot
plt.subplot(2,2,4)
plt.scatter(minutes, saves, color='yellow')
plt.ylabel('Saves')
plt.xlabel('Minutes')

plt.tight_layout()
plt.show()
```

```python
model = LinearRegression()
plt.figure(figsize=(10,6))

#model for passes
X_train, X_test, y_train, y_test = train_test_split(minutes, passes, random_state=42, test_size=0.3)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#passes plot
plt.subplot(2,2,1)
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.ylabel('Passes')
plt.xlabel('Minutes')

#model for shots
X_train, X_test, y_train, y_test = train_test_split(minutes, shots, random_state=42, test_size=0.3)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#shots plot
plt.subplot(2,2,2)
plt.scatter(X_test, y_test, color='green')
plt.plot(X_test, y_pred, color='red')
plt.ylabel('Shots')
plt.xlabel('Minutes')

#model for tackles
X_train, X_test, y_train, y_test = train_test_split(minutes, tackles, random_state=42, test_size=0.3)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#tackles plot
plt.subplot(2,2,3)
plt.scatter(X_test, y_test, color='purple')
plt.plot(X_test, y_pred, color='red')
plt.ylabel('Tackles')
plt.xlabel('Minutes')

#model for saves
X_train, X_test, y_train, y_test = train_test_split(minutes, saves, random_state=42, test_size=0.3)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

#saves plot
plt.subplot(2,2,4)
plt.scatter(X_test, y_test, color='yellow')
plt.plot(X_test, y_pred, color='red')
plt.ylabel('Saves')
plt.xlabel('Minutes')

plt.tight_layout()
plt.show()
```

3. Explain and apply Linear regression to predict ice cream revenue in iceream.csv

```python
ice_cream_df = pd.read_csv('IceCreamData.csv')

X = np.array(ice_cream_df[['Temperature']])
y = np.array(ice_cream_df[['Revenue']])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size = 0.3)

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

plt.scatter(X_test, y_test, color='green')
plt.plot(X_test, y_pred, color='red', linewidth=2)
plt.ylabel('Revenue')
plt.xlabel('Temperature')
plt.show()
```

*This shows a direct proportion between the temperature and the revenue of an ice cream. As the temperature goes up, the revenue also goes up.*


```python
df = pd.DataFrame({'Actual Values': y_test.flatten(), 'Predicted Values': y_pred.flatten()})
df.head(10)
```

```python
comparison_plot = df.head(10)
comparison_plot.plot(kind='bar', figsize=(5,5))
plt.grid(linestyle='-', linewidth='0.5', color='green')
plt.grid(linestyle=':', linewidth='0.5', color='red')
plt.title('Price value Comparison')
plt.show()
```

*This shows the values between actual and prediction, which is closer to each other.*




**Predicting Revenue based off of user's temperature input**

```python
rev = model.predict([[int(input("Input an ice cream temperature: "))]])
print(f'Predicted Revenue: {rev[0,0]}')
```

*Predicted revenue is somewhat closer to the actual value*
